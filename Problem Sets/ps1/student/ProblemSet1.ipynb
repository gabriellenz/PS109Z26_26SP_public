## Trends in Violence and Incarceration

Over the past half century, the United States has experienced a dramatic expansion of its incarceration system. Beginning in the late 1960s, and accelerating sharply in the 1980s, the number of people held in prisons and jails grew at an unprecedented pace. By the early 2000s, the U.S. incarceration rate far exceeded that of any other advanced democracy.

One of the most concerning features of this expansion is its timing. As many scholars have emphasized, incarceration continued to rise throughout the 1990s even as rates of crime and violence declined nationwide. In other words, the growth of the prison system did not simply track changes in public safety. Instead, incarceration increasingly became, according to these scholars, disconnected from underlying trends in crime.

In this problem set, we will examine this relationship more closely. Using national data on incarceration and violent crime, you will visualize how these trends evolved over time and assess whether changes in incarceration appear to move in tandem with changes in violence.
